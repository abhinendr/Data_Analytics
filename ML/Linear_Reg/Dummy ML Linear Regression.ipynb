```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
df_train.isnull().sum()
```

```python
df_train =df_train.dropna()
```

```python
df_train.shape
```

```python
sns.pairplot(df_train)
```

```python
plt.scatter(df_train[['x']],df_train.y)
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df_train[['x']],df_train.y) #x var should be 2d array
```

<b>Importing the test Module to be predicted

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df_test['Predicted_val'] = reg.predict(df_test[['x']])
```

```python
df_test.head()
```

<b>Check the accuracy of the model

```python
plt.title('Predicting the accuracy of the model')

plt.scatter(df_test.x, df_test.y, label='Original Data Model', c='Green')
plt.scatter(df_test.x, df_test.Predicted_val, label='Predicted Data Model', c='Red')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

<b>Checking the accuracy score

```python
Accuracy_score = reg.score(df_test[['y']],df_test[['Predicted_val']])
```
